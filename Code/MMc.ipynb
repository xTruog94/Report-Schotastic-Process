```python
import numpy as np
import math  
def fact(n):
    giai_thua = 1;
    if (n == 0 or n == 1):
        return giai_thua;
    else:
        for i in range(2, n + 1):
            giai_thua = giai_thua * i;
        return giai_thua;
def p0(c,k,l,m):
    r=l/m
    tong1=0
    tong2=0
    p0=0
    for i in range(c):
        tong1=tong1+pow(r,i)/fact(i);
    if r/c<1:
        p0=tong1+pow(r,c)/(fact(c-1)*(c-r))
    else:
        for i in range(c,k+1):
            tong2=tong2+pow(r,n)/(pow(c,n-c)*fact(c))
        p0=1/(tong1+tong2)
    return p0
def pn(c,k,l,m,n):
    r=l/m
    tong1=0
    tong2=0
    if n==0:
        for i in range(c):
            tong1=tong1+pow(r,i)/fact(i)
        for i in range(c,k+1):
            tong2=tong2+pow(r,i)/(pow(c,i-c)*fact(c))
        return 1 / (tong1+tong2)
    elif n<c:
        return pn(c,k,l,m,0)*pow(r,n)/fact(n)
    elif n <= k:
        return pn(c,k,l,m,0) * pow(r,n) / (pow(c,n-c) * fact(c))

#MMc  
    
def L(c,k,l,m):
    r=l/m
    L=0
    if c==1:
        if r==1:
            L=k/2
        else:
            for i in range(k+1):
                L=L+i*pow(r,i)*(1-r)/(1-pow(r,k+1))
    else:
        for i in range(k+1):
            L=L+i*pn(c,k,l,m,i)
    return L
def Lq(c,k,l,m):
    #c là số services
    #l=\lambda
    #m=\mu
    r=l/m;
    tong=0;
    Lq=0;
    if c==1:
        if r==1:
            return k * (k - 1) / 2 / (k + 1)
        else:
            return L(1,k,l,m) - r * (1 - pow(r,k)) / (1 - pow(r,k+1))
    else:
        for i in range(c,k+1):
            Lq=Lq+(i-c)*pn(c,k,l,m,i)
    return Lq;
#lamba_e=e

def W(c,k,l,m):
    e=l*(1-pn(c,k,l,m,k))
    W=L(c,k,l,m)/e
    return W
def Wq(c,k,l,m):
    e=l*(1-pn(c,k,l,m,k))
    Wq=Lq(c,k,l,m)/e
    return Wq
#xs chờ khi full dịch vụ
def wait(c,k,l,m):
    r=l/m
    tong=0
    for i in range(c,k+1):
        tong=tong+pn(c,k,l,m,i)
    return tong
#Xác suất từ chối
def Pr(c,k,l,m):
    r=l/m
    tong=0
    for i in range(k):
        tong=tong+pn(c,k,l,m,i)
    return 1-tong
#chi phí cơ hội
def c0(c,k,l,m,cost):
    c0=Pr(c,k,l,m)*l*cost
    return c0
def services(l,m,cq,cs):
    x=0
    x=(l+math.sqrt((cq*l*m)/cs))/m
    return x
services(16,20,4,20)
```
